```python
import pandas as pd
df=pd.read_csv('lung_cancer_examples.csv')
df
```

```python
df1=df.copy()
df1
```

```python
df1.drop(['Name','Surname'],axis=1,inplace=True)
```

```python
df1
```

```python
df1.isna().sum()
```

```python
df1.dtypes
```

```python
X=df1.iloc[:,:-1].values
X
```

```python
y=df1.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X)
X_scaled=scaler.transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
59*0.3
```

```python
59-18
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
y_test.values
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
y_new=knn.predict(scaler.transform([[30,0,5,2]]))
y_new
if y_new==0:
    print('person is not suffering lung cancer')
else:
    print('person is suffering lung cancer')
```
